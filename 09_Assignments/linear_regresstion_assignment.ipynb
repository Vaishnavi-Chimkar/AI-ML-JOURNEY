```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
house_price_Prediction = pd.read_csv("HousePricePrediction.csv")
house_price_Prediction.drop(['Id'],axis=1,inplace=True)
house_price_Prediction = house_price_Prediction.dropna()
house_price_Prediction = pd.get_dummies(house_price_Prediction,drop_first = True,dtype = int)
```

```python
X = house_price_Prediction.drop(columns = ["SalePrice"],axis =1)
y = house_price_Prediction["SalePrice"]

X["lotArea_overallCondn"] = X["LotArea"]*X["OverallCond"]

X.fillna(X.mean(),inplace = True)
y.fillna(y.mean(),inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)

SalesPrice_Prediction__model = LinearRegression()
SalesPrice_Prediction__model.fit(X_train,y_train)

y_pred = SalesPrice_Prediction__model.predict(X_test)
r2 = r2_score(y_pred,y_test)
mae = mean_absolute_error(y_test, y_pred)

print("r2 : ",r2)
print("mae : ",mae)
```
