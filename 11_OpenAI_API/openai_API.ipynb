AI Personal Assistant

```python
#queries answer 
#emails => summarize

class PersonalAssistant:
    def __init__(self):
        print("Hi, Iam your AI assistant. How can I help!")
    def ans_query(self):
        client = OpenAI(
            api_key=GROQ_API_KEY,
            base_url="https://api.groq.com/openai/v1",
        )

        question = input("Ask me anything: ")
        response = client.responses.create(
            input= [
                {"role":"system","content":"Act like a helpful personal assistant"},
                {"role":"user","content":question}
            ],
            model="openai/gpt-oss-20b",
            temperature=0.7,
            max_output_tokens=512
        )
        print(response.output_text.strip())

    def summarize_email(self):
        client = OpenAI(
            api_key=GROQ_API_KEY,
            base_url="https://api.groq.com/openai/v1",
        )

        email_text = input("Paste your email here: ")
        prompt = f"summarize the following email in 2-3 sentences: {email_text}"
        
        response = client.responses.create(
            input= [
                {"role":"system","content":"Act like a expert email assistant"},
                {"role":"user","content":prompt}
            ],
            model="openai/gpt-oss-20b",
            temperature=0.3,
            max_output_tokens=512
        )
        print("\n\nSummary: ",response.output_text.strip())
            
```

```python
assistant = PersonalAssistant()
```

```python
assistant.summarize_email()
```

```python
import base64
# images => base64 encoded strings => decode
```

```python
response = client.responses.create(
    input= "Generate an image of a highly futuristic spaceship. flying near a nebula",
    model="openai/gpt-oss-20b",
    tools = [{"type":"image_generation"}]
)

image_data = [
    output.result
    for output in response.output
    if output.type == "image_generation_call"
]

#decode & store image in a file

if image_data:
    img_base64 = image_data[0]
    with open("spaceship.png","wb") as f:
        image_bytes = base64.b64decode(img_base64)
        f.write(image_bytes)
```
